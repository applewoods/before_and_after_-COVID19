## 불용어 처리


```
import pandas as pd

news_2019 = pd.read_csv('./data/NaverNews2019.csv')
news_2020 = pd.read_csv('./data/NaverNews2020.csv')

stopwords = pd.read_csv('./data/stopwords.txt', '\t', header=None)
```

```
news_2019.head()
```

```
print(len(news_2019), len(news_2020))
```

```
news_2019['Body'].nunique(), news_2020['Body'].nunique()
```

```
# Body column 중복 값 제거
news_2019.drop_duplicates(subset=['Body'], inplace=True)
news_2020.drop_duplicates(subset=['Body'], inplace = True)
```

```
print(len(news_2019), len(news_2020))
```

```
# 한글과 공백을 제외하고 모두 제거
news_2019['Body'] = news_2019['Body'].str.replace("[^ㄱ-ㅎㅏ-ㅣ가-힣 ]","")
news_2020['Body'] = news_2020['Body'].str.replace("[^ㄱ-ㅎㅏ-ㅣ가-힣 ]","")

print(news_2019[:5], news_2020[:5])
```

```
import numpy as np
news_2019['Body'].replace('', np.nan, inplace=True)
news_2020['Body'].replace('', np.nan, inplace=True)

print(news_2019.isnull().sum(), news_2020.isnull().sum())
```

```
news_2019 = news_2019.dropna(how = 'any')
print(len(news_2019))
```

# 불용어 처리

```
stopwords = stopwords[0].values.tolist()
print(stopwords)
```

```
# 기존 불용어 외에 따로 추가할 불용어
stopwords_added = '에 을 과 며 인 를 및 등 은 의 다 와 은 본 서 로 는 구독 배포 머니투데이 리얼타임' 
stopwords_added = stopwords_added.split(' ')

stopwords.extend(stopwords_added)
```

```
print(stopwords)
```

```
from konlpy.tag import Okt
#from collections import Counter

okt = Okt()

X_2019 = []
for sent in news_2019['Body']:
    temp = []
    temp = okt.nouns(sent) # 토큰화
    temp = [word for word in temp if not word in stopwords]
    X_2019.append(temp)
  
X_2020 = []
for sent in news_2020['Body']:
    temp = []
    temp = okt.nouns(sent) # 토큰화
    temp = [word for word in temp if not word in stopwords]
    X_2020.append(temp)
```

```
X_2019
```

```
count_2019 = Counter(X_2019)
```

```
noun_2019 = count_2019.most_common()
for _ in noun_2019:
  print(_)
```

```
noun_2020 = count_2020.most_common()
for _ in nount_2020:
  print(_)
```
